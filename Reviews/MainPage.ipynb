```python
import requests as rq
from bs4 import BeautifulSoup
import pandas as pd
```

```python
p_list=[]
```

```python
def parseData(link):
    page=rq.get(link)
    soup=BeautifulSoup(page.content,'lxml')
    plist = soup.find('div',class_='attraction_list attraction_list_short ')
    places = plist.find_all('div',class_='attraction_element')
    for p in places:
        tname=p.find('div',class_='listing_title').a.text
        tlink=p.find('div',class_='listing_title').a['href']
        trating=p.find('div',class_='rs rating').span['alt']
        tnreviews=p.find('span',class_='more').text
        tcategory=p.find('div',class_='p13n_reasoning_v2').span.text
        tplace={'name':tname,'link':tlink,'rating':trating,'reviews':tnreviews,'category':tcategory}
        p_list.append(tplace)
    nextpage=plist.find('a',class_='nav next rndBtn ui_button primary taLnk')
    gotonext='https://www.tripadvisor.in'+nextpage['href']
    return(gotonext)
```

```python
link='https://www.tripadvisor.in/Attractions-g297604-Activities-Goa.html'
xyz=parseData(link)
```
