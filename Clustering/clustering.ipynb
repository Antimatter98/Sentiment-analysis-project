```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('D:\\testfinal2.xlsx')
```

```python
lst = []
```

```python
lst = df.category.unique()
```

```python
lst = np.ndarray.tolist(lst)
```

```python
df2 = pd.DataFrame({'weight':df.groupby(['name','category'])['ratings'].mean()}).reset_index()
```

```python
cat_ind = dict(list(zip(lst,range(0,len(lst)))))
```

```python
names = df2.name.unique()
```

```python
all_fv = []
for name in names:
    tfv = [0]*len(lst)
    temp = df2[df2['name']==name]
    cat_weights = dict(zip(temp['category'],temp['weight']))
    for t in temp['category']:
        ind = cat_ind[t]
        tfv[ind] = cat_weights[t]
    all_fv.append(tfv)
```

```python
arr = np.asarray(all_fv)
```

```python
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt
```

```python
def augmented_dendrogram(*args, **kwargs):

    ddata = dendrogram(*args, **kwargs)

    if not kwargs.get('no_plot', False):
        for i, d in zip(ddata['icoord'], ddata['dcoord']):
            x = 0.5 * sum(i[1:3])
            y = d[1]
            plt.plot(x, y, 'o')
            plt.annotate("%.3g" % y, (x, y), xytext=(0, -5),
                         textcoords='offset points',
                         va='top', ha='center')

    return ddata
```

```python
plt.title('Agglomerative Clustering')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
augmented_dendrogram(
    Z,
    truncate_mode='lastp',  
    p=8,  
    leaf_rotation=90,
    leaf_font_size=9,
    #show_contracted=True,  
)
plt.show()

```
